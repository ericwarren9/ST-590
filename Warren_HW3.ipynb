### Part 1

 This question doesn't involve any programming. Suppose you have a pandas `DataFrame`. In a markdown cell, describe what each of the following operators/methods allows you to subset from a data frame (that is, describe all of the types of subsetting you can do with the operator):
- []
- .iloc[]
- .loc[]

- By using the [] method, we can return **either** row(s) **or** column(s). However, we cannot return both at the same time. To get the column(s) we need to do something like `df["column_name"]` or to get multiple we do `df[["column_name1", "column_name2", ...]]`. To get the row(s) we do it via slicing using the integer of what row number(s) we want. We must have at least more than one value present meaning that if we want the first row we cannot do `df[0]` and must do `df[0:1]`.
- By using the .iloc[] method, we can return **both** rows **and** columns. The rows and columns specified must be in integer format and the format of how we would subset via this method is `df.iloc[rows, columns]`. An example to get the first two rows and two columns would be `df.iloc[0:2, 0:2]`.
- By using the .loc[] method, we can return **both** rows **and** columns. The rows and columns specified can be in integer **or** name (or label) format. Note that subsetting in .loc[] includes the last value in slicing, which is different than the others. An example of we could use .loc[] is `df.loc[0:5, "column_name"]` which gives us the first **six** rows and the column "column_name".
- Note we can use all of these to subset with booleans using logicals.

First we will look at the one-way table. We are going to look at the number of each sex of student.

```python
student_data_math.sex.value_counts()
```

From here, we can see that we have slightly more females with them outnumbering the number of male students by 208 to 187.

Now we are going to look at a two way contingency table looking at the sex of a student and their family size.

```python
pd.crosstab(student_data_math.sex,
            student_data_math.famsize)
```

There are a couple things to note. The first is that most students have a family size greater than 3, meaning that on top of their parents they have at least one other sibling. Now we are going to look at a three way contingency table looking at the sex of a student, their family size, and school.

```python
pd.crosstab([student_data_math.sex, student_data_math.famsize],
            student_data_math.school)
```

On top of what we have seen before, we can clearly see that more math students go to Gabriel Pereira than Mousinho da Silveira being the other school. It seems like the proportion of male students to female students seem to be about the same at each school and same with the family sizes.
